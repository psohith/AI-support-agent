```python
import pandas as pd
import google.generativeai as genai
```

```python
genai.configure(api_key="AIzaSyBo2HlMw0AvGEeV8dS6gjoa3bsO7MGMGi0")
model = genai.GenerativeModel("models/gemini-2.0-flash-exp")
```

```python
prompt_template = """
Refactor the following text to make it concise, clear, and structured. Remove redundancies, streamline explanations, and present the information in a professional and user-friendly tone, using bullet points or short paragraphs for clarity. Keep all relevant details intact:

{text}

Ensure the result has the following characteristics:
- Uses proper grammar and punctuation.
- Is easy to read and understand.
- Summarizes processes in a step-by-step format when applicable.
- Includes important headings or subheadings for clarity.
- Avoids any introductory or concluding remarks; provide only the refactored summary.
"""
```

```python
input_file = "/content/scraped_content.csv"
output_file = "/content/summarized_content.csv"
data = pd.read_csv(input_file)
data.head()
```

```python
data.isnull().sum()
```

```python

def summarize_content(content):
    if pd.isna(content) or not content.strip():
        return ""
    prompt = prompt_template.format(text=content)
    try:
        response = model.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        print(f"Error processing content: {e}")
        return "Error during summarization"
```

```python
data["summarized_text"] = data["Content"].apply(summarize_content)
```

```python
data.to_csv(output_file, index=False)
print(f"Summarized content added and saved to {output_file}")
```
